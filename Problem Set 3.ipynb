```python
import pandas as pd
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|', index_col='user_id')

MeanAge = users.groupby('occupation')['age'].mean()
print("Step 4: Discover what is the mean age per occupation:")
print(MeanAge)

users['male'] = (users['gender'] == 'M').astype(int)

MaleRatio = users.groupby('occupation')['male'].mean()
MaleRatio.sort_values(ascending=False, inplace=True)
print("Step 5: Discover the male ratio per occupation:")
print(MaleRatio)


MinMaxAge = users.groupby('occupation')['age'].agg(['min', 'max'])
print("Step 6: For each occupation, calculate the minimum and maximum ages:")
print(MinMaxAge)

MeanAge_O_S = users.groupby(['occupation', 'gender'])['age'].mean()
print("Step 7: For each combination of occupation and sec, calculate the mean age:")
print(MeanAge_O_S)


GenderCount = users.groupby(['occupation', 'gender'])['gender'].count()
OccupationCount = users.groupby('occupation')['gender'].count()
GenderPercentage = (GenderCount / OccupationCount) * 100
print("Step 8: For each occupation preseent the percentage of women and men:")
print(GenderPercentage)

```

# Question No 4

```python
import pandas as pd
import numpy as np
from datetime import datetime
filename = 'wind.txt'
data=pd.read_csv('wind.txt',sep="\s+",parse_dates=[[0,1,2]])
data.head()
data.columns = ['date', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR',
                'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']

data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

data.head()
def fix_year(date):
    year = date.year - 100 if date.year > 1989 else date.year
    return datetime.datetime(year, date.month, date.day)

data.head()
print("Number of missing values for each location:")
print(data.isnull().sum())
print("Total number of non-missing values:")
print(data.notnull().sum().sum())
print("Mean windspeed over all locations and times:")
print(data.mean().mean())

loc_stats = pd.DataFrame()
loc_stats['min'] = data.min()
loc_stats['max'] = data.max()
loc_stats['mean'] = data.mean()
loc_stats['std'] = data.std()

print("Statistics for each location:")
print(loc_stats)

day_stats = pd.DataFrame()
day_stats['min'] = data.min(axis=1)
day_stats['max'] = data.max(axis=1)
day_stats['mean'] = data.mean(axis=1)
day_stats['std'] = data.std(axis=1)

print("Statistics for each day:")
print(day_stats)

january_data = data[data.index.month == 1]
january_means = january_data.mean()
print("Average windspeed in January for each location:")
print(january_means)

yearly_data = data.resample('AS').mean()
print("Yearly windspeed data for each location:")
print(yearly_data)

monthly_data = data.resample('MS').mean()
print("Monthly windspeed data for each location:")
print(monthly_data)

data.resample('W').mean()

weekly_data = data.resample('W', closed='left', label='left').mean().iloc[1:53]
weekly_stats = weekly_data.agg(['min', 'max', 'mean', 'std'], axis=1)
print(weekly_stats)
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipotle = pd.read_csv(url, sep='\t')


print(chipotle[:10])


print("The number of observations in the dataset is", len(chipotle))
print("The number of columns in the dataset is", len(chipotle.columns))
print("The name of all the columns are:", chipotle.columns)


dataset = chipotle[["item_name", "quantity", "choice_description", "item_price"]]


most_ordered_item = chipotle.groupby('item_name')['quantity'].sum().sort_values(ascending=False).index[0]
print("The most-ordered item is:", most_ordered_item)


items_ordered = chipotle.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[0]
print("The number of items ordered for the most-ordered item is:", items_ordered)


most_ordered_item_choice = chipotle.groupby('choice_description')['quantity'].sum().sort_values(ascending=False).index[0]
print("The most-ordered item in the choice_description column is:", most_ordered_item_choice)


total_items_ordered = chipotle['quantity'].sum()
print("The total number of items ordered is:", total_items_ordered)


chipotle['item_price'] = chipotle['item_price'].str.replace('$', '', regex=True).astype(float)
print("The item_price type after conversion is:", chipotle['item_price'].dtypes)



revenue = (chipotle['quantity'] * chipotle['item_price']).sum()
print("The revenue for the period in the dataset is: $", round(revenue, 2))


orders = chipotle['order_id'].nunique()
print("The total number of orders made in the period is:", orders)


avg_revenue = round(revenue / orders, 2)
print("The average revenue amount per order is: $", avg_revenue)


items_sold = chipotle['item_name'].nunique()
print("The number of different items sold is:", items_sold)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

data['Marriage_Rate'] = data['Marriages'] / data['Population']
data['Divorce_Rate'] = data['Divorces'] / data['Population']

plt.plot(data['Year'], data['Marriage_Rate'], label='Marriages')
plt.plot(data['Year'], data['Divorce_Rate'], label='Divorces')

plt.xlabel('Year')
plt.ylabel('Per Capita Rate')
plt.title('Marriage and Divorce Rates in the U.S., 1867-2014')
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('actor_kill_counts.csv')
data['Actor'] = data['Actor']
data['Kill Count'] = data['Count']
df = pd.DataFrame(data)

df = df.sort_values(by='Kill Count', ascending=True)


ax = df.plot(kind='barh', x='Actor', y='Kill Count')

ax.set_title('Deadliest Actors in Hollywood')
ax.set_xlabel('Kill Count')
ax.set_ylabel('Actor')


for i, v in enumerate(df['Kill Count']):
    ax.text(v + 3, i, str(v), color='black', fontweight='bold')

plt.show()

```

# Question No 9

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('roman-emperor-reigns.csv')
TotalEmperors = len(data)
Assassinated = len(data[data['Cause_of_Death'] == 'Assassinated'])
AssassinatedPercentage = (Assassinated / TotalEmperors) * 100
SurvivedPercentage = 100 - AssassinatedPercentage

labels = ['Assassinated', 'Survived']
sizes = [AssassinatedPercentage, SurvivedPercentage]
colors = ['pink', 'orange']

plt.pie(sizes, labels=labels, colors=colors, autopct='%3.3f%%', startangle=90)
plt.title('Assassinated Roman Emperors')
plt.axis('equal')


plt.show()
```
